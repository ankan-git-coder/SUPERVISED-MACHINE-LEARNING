```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
X = np.array([[0.5], [1.0], [1.5], [2.0], [2.5], [3.0], [4.0], [5.0]])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model= KNeighborsClassifier(n_neighbors=3)
model.fit(X_scaled, y)
```

```python
new_student=np.array([[2.2]])
new_student_scaled = scaler.transform(new_student)
```

```python
prediction = model.predict(new_student_scaled)
probability = model.predict_proba(new_student_scaled)

print(f"Prediction for 2.2 hours: {prediction}")
print(f"Confidence (Probabilities): {probability}")
```
